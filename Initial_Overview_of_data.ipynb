## Important features to look at in the data

- gameId / playId - tie into the 2 previous dataframes

- playdescription - seems less useful but will be good to keep to have an idea of the play in a much more readable way

- quarter/down/yards to go - all important when trying to analyze the plays. longer yardstogo, on 2nd/3rd/4th down will likely result in less successful but deeper passes and vice versa for shorter yardstogo.

- playtype are all passing plays. Some will be sacks, some passes and unknown will encompass the rest, likely interceptions/flags that result in yards, all important as they are the results. May not be included when trying to analyze results in a model. Will be very correlated to whether or not a play is successful

- yardslineside / yardlinenumber - where they are starting the play. very important. bigger fields usually lead to more successful chances and more points

- offenseFormation / personnelO - the way the offense is set up and how many of each position they have. Will need to clean this up as there are some weirdly set up formations that sometimes write 0 of a position, as well as have defensvive formations in there.

- defendersInTheBox / numberofPassRushers - overview of the type of pressure the other team is putting on the offense.

- personnelD - same as offensive personnel for the other side. again needs to be cleaned so it doesnt include defensive players

- typeDropback - the type of passing play. Very important as scrambles are often a biproduct of aggressive defense and can result in big plays or busts, more rare to be inbetween.

- 'preSnapVisitorScore', 'preSnapHomeScore', 'gameClock' - overview of the current state of the game. similar to down and distance could be very valuable as teams further behind will look for bigger plays while teams ahead may be safer

- absoluteyardlinenumber - can be covered using the yardline numbers and side instead. but could be useful to see how deep into the endzone some plays end up

- penaltyCodes - due to the excessiveness of the distinct types of penalties will likely turn out to be much less useful than the yardage result

- penaltyJerseyNumbers - who commited the penalty

- pass result - catch, incomplete, sack, interception (r is likely run but that is not relevant to us)

- offensive play result / play result - these are the be all end all of this. how many yards does each play get. The big distinction is that offensive play result does not include penalties, just the actual catch/drop. where as play result is the final action including penalties. Will be checked and if they are not equal it means there was a penalty and if they are then there was no penalty or the penalty was declined.

- epa - estimated points added

- is defensive past interference - important penalty for passing plays

## Understanding the data

- time - time stamp of the play, time, y-m-d, h-m-s
- x - players x position on the field, goes from 0-120
- y - players y position on the field, goes from 0 to 53.3
- s - spead in yards per second
- a - acceleration in yards per second squared
- dis - distance traveled from the previous time point
- o - player orientation - in degrees 0-360
- dir - angle of players motion in degrees 0-360
- event - tagged play details (for moment of ball snap, pass release, pass catch etc
- nflId - player identification (ties in to player data)
- display name - player name
- jersey number - player number
- position - player position group
- team - home or away of corresponding player
- frameid - frame identifier for each play
- gameid - correlates to the games dataframe
- playid - game identifier
- play direction - left or right
- route - route ran by a specific player

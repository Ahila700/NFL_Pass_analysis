The data is organized on a play by play basis. So for each play it will give the positional data of every single player on that play involved in it. This will likely need to be tied into the plays dataframe also given. 

PlayId == 75 was a 15 yard pass play to julio jones. So using that data and this one we can see that julio jones ran a hitch route where the player makes a hard sprint then turns and stops. So we can see based on the y location that julio jones was being guarded by jalen mills. And jalen mills gave him an 8 yard cushion on the snap. Then right as the pass was being thrown he had an acceleration of 6.67 yards/s^2 where jalen mills waiting for him had a much slower speed of 2.18. Then as the ball is arriving we see that jalen mills begins to speed up to match it where as julio, running a hitch slows down instead. And creates more distance when the pass arrives then when it is thrown. 

These are factors that will weigh heavily on a model to see if a pass is completed or not. For example on a hitch route taking data like the speed of the players and the distance between them on the pass forward and the pass arrival are very important as both give a good idea of how much room the receiver has to make the completion. 

Even on this play Sanu, running the same route does not get the ball and when you look at his acceleration/speed and distance to his defender malcolm jenkins you can see why. Jenkins is all over him on the pass forward and moving at a faster acceleration. So having less space will mean he is likely not to get the ball. If Julio jones was covered the same way as the only 2 real passing threats this will have likely led to an incomplete pass or even possibly a sack. So those are the factors that will need to be looked at further.
